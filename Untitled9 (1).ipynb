```python
import pandas as pd
import os
from datetime import datetime as dt
import math
import warnings
warnings.filterwarnings('ignore')

tdata=pd.read_csv(r"C:\Users\SSVD HEMANTH\Documents\hemanth\project-final\Copy of MALAMPUZHA_Water_Extent_hemanth.csv")
tdata.head()
```

```python
tdata['Date'] = pd.to_datetime(tdata['Date'])
#tdata.groupby(tdata.index.Date).Volume.sum().reset_index()
tdata=tdata.ffill()

tdata['day'] = tdata['Date'].dt.day
tdata['month'] = tdata['Date'].dt.month
tdata['year'] = tdata['Date'].dt.year

x=tdata[['day','month','year']]
y=tdata['CRL(m)']
y
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(y, period=4)
result.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(tdata['Area (km2)'])
plot_pacf(tdata['CRL(m)'])
```

```python
from statsmodels.tsa.stattools import adfuller
adf, pVal, usedLag, nobs, critVals, icBest =  adfuller(y)
print('ADF test statistic:', adf)
print('ADF p-values:', pVal)
print('ADF number of lags used:', usedLag)
print('ADF number of observations:', nobs)
print('ADF critical values:', critVals)
print('ADF best information criterion:', icBest)
```

```python
import matplotlib.pyplot as plt
def plotPrediction(yTrain, yTest, yPred):
    xAxis1 = yTest.index
    plt.title('CRL Plot')
    plt.xlabel('')
    plt.ylabel('CRL')
    plt.plot(xAxis1, yTest, label='Real Data')
    plt.plot(xAxis1, yPred, label='Prediction')
    plt.legend(loc='lower right')
    plt.show()
    xAxis2 = yTrain.index
    plt.title('CRL Plot')
    plt.xlabel('')
    plt.ylabel('CRL')
    plt.plot(xAxis2, yTrain, label='Real Training Data')
    plt.plot(xAxis1, yPred, label='Prediction')
    plt.legend(loc='lower right')
    plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def calcMetrics(yTest, yPred, modelName):
    rmse = round(mean_squared_error(yTest, yPred, squared=False), 5)
    mae = round(mean_absolute_error(yTest, yPred), 5)
    r2 = round(r2_score(yTest, yPred), 5)
    return (rmse, mae, r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = sqrt(mse_rf)
mae = round(mean_absolute_error(y_test, y_pred_rf), 5)
r2 = round(r2_score(y_test, y_pred_rf), 5)
print(f'Random Forest RMSE: {rmse_rf}')
plotPrediction(y_train, y_test, y_pred_rf)
print('MAE:', mae)
print('R2 Score:', r2)
```

```python
from xgboost import XGBRegressor
regg = XGBRegressor()
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
regg.fit(X_train, y_train)
y_pred_rf = regg.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = sqrt(mse_rf)
mae = round(mean_absolute_error(y_test, y_pred_rf), 5)
r2 = round(r2_score(y_test, y_pred_rf), 5)
print(f'Random Forest RMSE: {rmse_rf}')
plotPrediction(y_train, y_test, y_pred_rf)
print('MAE:', mae)
print('R2 Score:', r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regg = GradientBoostingRegressor()
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
regg.fit(X_train, y_train)
y_pred_rf = regg.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = sqrt(mse_rf)
mae = round(mean_absolute_error(y_test, y_pred_rf), 5)
r2 = round(r2_score(y_test, y_pred_rf), 5)
print(f'Random Forest RMSE: {rmse_rf}')
plotPrediction(y_train, y_test, y_pred_rf)
print('MAE:', mae)
print('R2 Score:', r2)
```
